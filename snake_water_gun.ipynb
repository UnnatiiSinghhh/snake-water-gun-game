```python
import random  #Module which chooses random values

items=["snake","water","gun"]

Tie="It's a tie!"
Lose="You lose!"
Win="You won!"
print("Hello, welcome\nSnake-Water-Gun is a fun 2-player game, where each player chooses one of these three items:\nðŸ Snake\nðŸ’§ Water\nðŸ”« Gun")
print("Each item can defeat one of the others, and can be defeated by one.ðŸŽ®-")
print("1- SnakeðŸ drinks waterðŸ’§ â†’ SnakeðŸ wins\n2- WaterðŸ’§ drowns gunðŸ”« â†’ WaterðŸ’§ wins\n3- GunðŸ”« shoots snakeðŸ â†’ GunðŸ”« wins\n4- If both choose the same â†’ Tie")

while True:
    while True:
        total_rounds=int(input("Enter the number of rounds you want to play: "))
        if(total_rounds>=1 and total_rounds<=15):
            break
        else:
            print("Enter valid number of rounds.\nThe limit is 15")

    round=1
    score=0
    py_score=0
    game_quit=False
    skip_result=False

    while(round<=total_rounds):
        print(f"Round {round}")
        user_choice=input("Enter your choice(snake, water or gun??): ").lower()

        if user_choice in items:
            py_choice=random.choice(items)
            print(f"Computer chose: {py_choice}")

            if(py_choice==user_choice):
                print(Tie)
                
                if(total_rounds==1):
                    print("Do you want to play again??")
                    
                    while True:
                        ask = input("yes/no: ").lower()
                        if ask == "yes":
                            skip_result=True
                            break  # Restart the game
                        elif ask == "no":
                            print("Game stopped early! Thank you for playing.")
                            game_quit = True
                            break  # Exit replay question loop
                        else:
                            print("Please enter a valid option: 'yes' or 'no'")
                    if game_quit:
                        break  #End the game
                    if not skip_result:  #This will not print result if player chooses to play again after a tie in the one and only round
                        print(f"Your score is {score} and computer's score is {py_score}")
                
            elif(py_choice=="snake" and user_choice=="water") or (py_choice=="water" and user_choice=="gun") or (py_choice=="gun" and user_choice=="snake"):
                print(Lose)
                py_score+=1
                print(f"Your score is {score} and computer's score is {py_score}")
            else:
                print(Win)
                score+=1  
                print(f"Your score is {score} and computer's score is {py_score}")
        
            round+=1
        
        else:
            print("Choose a valid option: snake, water or gun?\nLet's try again!")  

    if game_quit:
        break  #Stops the whole game early and doesn't shows score if it's a tie and player chooses not to play again.

    if skip_result:
        continue #Restart the game loop if player chooses to play again, without printing scores

    if(py_score>=score):
        a=py_score-score
        print(f"Computer won by {a}, better luck next timeðŸ’ªðŸ˜Š")
    elif(score>=py_score):
        b=score-py_score
        print(f"CongratulationsðŸŽ‰ðŸŽ‰!! You won by {b}")
    else:
        print("It's a tie!")
        
    print("Do you want to play again??")
    while True:
        ask = input("yes/no: ").lower()
        if ask == "yes":
            break  # Restart the game
        elif ask == "no":
            print("Game stopped early! Thank you for playing.")
            game_quit=True
            break  # Exit replay question loop
        else:
            print("Please enter a valid option: 'yes' or 'no'")
            continue  # Restart the replay question loop
    if not game_quit:
        continue  #Restart the game(game logic loop)
    else:
        break  #Exit the game after printing final scores
```
